```python
df['Heart Risk Score'] = (
    (df['Cholesterol'] > 200).astype(int) +
    df['Diabetes'] +
    df['Family History'] +
    df['Smoking'] +
    (df['BMI'] > 30).astype(int) +
    (df['Stress Level'] > 7).astype(int)
)

# Add a new column for Risk Score (simplified for visualization)
df['Risk Score'] = (
    df['Smoking'] +
    df['Obesity'] +
    (df['Alcohol Consumption'] > 8).astype(int) +
    (df['Sedentary Hours'] > 8).astype(int) +
    (df['Diet'] == 'Unhealthy').astype(int)
)
```

```python
# use the Smoth in which classes has equel representation

from imblearn.over_sampling import SMOTE
smote=SMOTE(random_state=365)
x_smote,y_smote=smote.fit_resample(x,y)
```

```python
#Chect the values counts of y smote

y_smote.value_counts(normalize=True)
```

```python
# Again split the data on y smote and x smote

x_train,x_test,y_train,y_test=train_test_split(x_smote,y_smote, test_size=0.2)
```

#13. Mengimplementasi sebuah Rabust Scaler yang akan membantu dalam proses Machine Learning

```python
# scale the vaues of cols usin Rabust Scaler

from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
scaler.fit(x_train[cols_to_scale])
x_train[cols_to_scale]=scaler.transform(x_train[cols_to_scale])
x_test[cols_to_scale]=scaler.transform(x_test[cols_to_scale])
```

#14. Melakukan Machine Learning model dalam Mengklasifikasi Data dalam semua Data

```python
# Import the modules
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#train gussain Model
clf=GaussianNB()
clf.fit(x_train,y_train)
clf_score=clf.score(x_test,y_test)
print(clf_score)
```

```python
# Logistic Regression
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(x_train, y_train)
y_pred_lr = lr.predict(x_test)
print(accuracy_score(y_test, y_pred_lr))
```

```python
# Get prediction from these model
#store the key valuies of models
models=[clf,lr]
model_prediction={}
for model in models:
    model_name=str(model).split("(")[0]
    model_prediction[model_name]=model.predict(x_test)
    print(model_prediction)
```

```python
logreg = LogisticRegression(max_iter=10000, random_state=42)
logreg.fit(x_train, y_train)

y_pred_logreg = logreg.predict(x_test)

print("\nLogistic Regression Report:")
print(classification_report(y_test, y_pred_logreg))

print("Accuracy:", accuracy_score(y_test, y_pred_logreg))

print("=================================================================================================================================")
print("=================================================================================================================================")

plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred_logreg)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix Logistic Regression')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
g = GaussianNB()
g.fit(x_train, y_train)

y_pred_g = g.predict(x_test)

print("\nNaye base Classifier:")
print(classification_report(y_test, y_pred_g))

print("Accuracy:", accuracy_score(y_test, y_pred_g))

print("=================================================================================================================================")
print("=================================================================================================================================")

plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test, y_pred_g)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Naye base Classifier')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

#15. Semua Hasil Dalam Evaluasi

```python
print("\n=== Model Comparison Semua Data===")
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_logreg))
print("Naye base Classifier:", accuracy_score(y_test, y_pred_g))
```

```python
# Evaluate the perforamnce model
from sklearn.metrics import accuracy_score,\
                                           f1_score, precision_score,recall_score

model_evaluation={}
for model, preds in model_prediction.items():
    model_evaluation[model]=[
        round(accuracy_score(y_test,preds)*100,2),
        round(f1_score(y_test,preds)*100,2),
        round(precision_score(y_test,preds)*100,2),
        round(recall_score(y_test,preds)*100,2),
    ]
```

```python
#Get the results
results_df=pd.DataFrame(model_evaluation, index=["Accuracy","F1 Score","Precision","recall"])
results_df.style.background_gradient(axis=None, cmap="bone")
```

```python
# Data akurasi untuk masing-masing model
model_names = [
    "Logistic Regression",
    "Naive Bayes"
]

# Akurasi prediksi model
model_accuracies = [
    accuracy_score(y_test, y_pred_logreg),
    accuracy_score(y_test, y_pred_g)
]

# Plotting bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(model_names, model_accuracies, color='skyblue', edgecolor='black')

# Annotating bar values
for bar in bars:
    plt.text(
        bar.get_width() + 0.01,
        bar.get_y() + bar.get_height() / 2,
        f"{bar.get_width():.2f}",
        va='center'
    )

# Adding titles and labels
plt.title("Model Comparison: Accuracy Scores", fontsize=14)
plt.xlabel("Accuracy", fontsize=12)
plt.ylabel("Models", fontsize=12)
plt.xlim(0, 1)
plt.grid(axis='x', linestyle='--', alpha=0.6)

# Show plot
plt.tight_layout()
plt.show()
```
